# Translating Stellar abundances to planet-equivelent interior 

Stellar abundances, specifically refractory element ratios, can be translated into equivalent planetary interiors by assuming the planet and its host star share these fundamental proportions. When defining these interiors, the choice of the stellar prior is a critical step. By default, we assume that the planet's bulk Fe/Mg and Fe/Si ratios match those of the star. However, our chemical model also supports the inclusion of minor mantle minerals (such as Ca- and Al-bearing phases) and allows for variable amounts of Si-Ni alloy in the core or Fe in the mantle. Therefore, if stellar abundances for Ni, Ca, or Al are available, the prior is expanded to include Ni/Fe, Ca/Mg, and Al/Mg ratios, thereby encompassing a more comprehensive range of interior possibilities.

```python
# import relevant modules
import exopie
import numpy as np

# Initialize stellar abundances normalized to the sun (assuming Asplund 2021)
N = 5000
Fe = np.random.normal(0,0.1,N)
Mg = np.random.normal(0,0.1,N)
Si = np.random.normal(0,0.1,N)
Ni = np.random.normal(0,0.1,N)
Al = np.random.normal(0,0.1,N)
Ca = np.random.normal(0,0.1,N)
# Set boundary on xSi and xFe (defualt is [0,0.2])
xSi = [0,0.2]
xFe = [0,0.2]
# Initialize star parameters
star = exopie.star(Fe=Fe,Mg=Mg,Si=Si,Ni=Ni,Ca=Ca,Al=Al,N=N,xSi=xSi,xFe=xFe)
# Running the model by passing the star data
star.to_planet()
```

However, recent empirical studies exploring star-planet chemical links suggest that this relationship may not be strictly 1-to-1. To account for this, our framework allows users to adjust the stellar prior to accurately reflect these newly derived scaling relations.

```python
# For example if the refractory ratios of planets have a linear relationship with host star
# e.g. Fe/Mg_planet = Fe/Mg_star*2+0.1
# We can modify the prior to be as following
prior=['Fe/Mg*2+0.1','Mg/Si']
star_new = exopie.star(Fe=Fe,Mg=Mg,Si=Si,N=N,prior=prior)
star_new.to_planet()
```
